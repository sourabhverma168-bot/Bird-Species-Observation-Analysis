````python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
````

### Dataset Loading

````python
grassland_url ="https://github.com/sourabhverma168-bot/Bird-Species-Observation-Analysis/blob/main/Bird_Monitoring_Data_GRASSLAND.XLSX?raw=true"
forest_url ="https://github.com/sourabhverma168-bot/Bird-Species-Observation-Analysis/blob/main/Bird_Monitoring_Data_FOREST.XLSX?raw=true"

# This list will hold every single sheet from both files before we merge them
all_data_frames = []

grassland_excel = pd.read_excel(grassland_url, sheet_name=None)

for sheet_name, df in grassland_excel.items():
    df['Source_Ecosystem'] = 'Grassland'  # Tag the ecosystem
    df['Source_Sheet'] = sheet_name       # Tag the specific park (e.g., 'WOTR')
    all_data_frames.append(df)

    # ==========================================
# 3. Download and Process FOREST Data
# ==========================================

forest_excel = pd.read_excel(forest_url, sheet_name=None)

for sheet_name, df in forest_excel.items():
    df['Source_Ecosystem'] = 'Forest'     # Tag the ecosystem
    df['Source_Sheet'] = sheet_name       # Tag the specific park
    all_data_frames.append(df)

    # ==========================================
# 4. Combine Everything into One Dataset
# ==========================================
print("Merging all 22 sheets together...")
final_df = pd.concat(all_data_frames, ignore_index=True)

# 5. Show the final results!
print(f"\nSUCCESS! Your consolidated dataset has {final_df.shape[0]} rows and {final_df.shape[1]} columns.")

````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
print("Starting Data Cleaning Process...")

# 1. Combine the two Taxon Code columns
# This says: "If TaxonCode is blank, fill it with whatever is in NPSTaxonCode"
final_df['TaxonCode'] = final_df['TaxonCode'].fillna(final_df['NPSTaxonCode'])

# Now that they are combined, we can delete the extra NPSTaxonCode column!
final_df = final_df.drop(columns=['NPSTaxonCode'])
print("✅ Taxon Codes successfully merged!")

# 2. Fill in the missing Site Names for the Grassland data
final_df['Site_Name'] = final_df['Site_Name'].fillna('Not Recorded')
print("✅ Missing Site Names patched!")

# 3. Clean up the Bird Details (Sex)
# If the observer didn't record the sex, we will label it 'Undetermined'
final_df['Sex'] = final_df['Sex'].fillna('Undetermined')
print("✅ Blank Sex columns filled!")

# 4. Final Sanity Check
print("\n--- Remaining Missing Data ---")
# This will show us if there are any other columns that still have blanks
remaining_missing = final_df.isnull().sum()
print(remaining_missing[remaining_missing > 0])
````

### What all manipulations have you done and insights you found?

**Data Manipulations Performed:**

1. Column Consolidation: Merged NPSTaxonCode into TaxonCode to create a single, unified species ID column, and dropped the redundant column.

2. Missing Value Imputation (Location): Filled the blank Site_Name cells (which occurred because Grassland data didn't track this) with the placeholder "Not Recorded" to prevent errors in future location-based analysis.

3. Missing Value Imputation (Biological): Replaced blank entries in the bird Sex column with "Undetermined".

**Insights Found:**
1. Data Template Inconsistencies: The primary insight is that the Forest and Grassland observers used slightly different data collection templates (mismatched Taxon column names and missing Site Names).

2. Field Observation Realities: Observers frequently cannot determine the sex of a bird in the wild (often due to distance or quick flyovers), making "Undetermined" a necessary and valid analytical category rather than just "missing data."

3. Dataset Harmonization: By fixing these structural mismatches, the two distinct datasets are now properly aligned into one cohesive table, allowing for accurate side-by-side comparisons of Forest vs. Grassland biodiversity in the next phase!

#### Chart - 5

````python
# Chart - 5 visualization code
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd

print("Generating Improved Vulnerability Matrix (Dark Mode)...")

# 1. Map disturbance to numerical scores so we can average them
dist_map = {
    'No effect on count': 0,
    'Slight effect on count': 1,
    'Moderate effect on count': 2,
    'Serious effect on count': 3
}
final_df['Dist_Score'] = final_df['Disturbance'].map(dist_map)

# 2. Safely count At-Risk birds
def count_vulnerable(x):
    return x.isin([True, 1, 'True', 'At-Risk (Watchlist)']).sum()

q1_df = final_df.groupby('Source_Sheet').agg(
    At_Risk_Count=('PIF_Watchlist_Status', count_vulnerable),
    Avg_Disturbance=('Dist_Score', 'mean')
).reset_index()

# Drop any potential NaN values from disturbance and sort highest to lowest
q1_df = q1_df.dropna().sort_values(by='At_Risk_Count', ascending=True)

# 3. Create a Horizontal Bar Chart (Setup for Dark Mode)
fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('black') # Outer background
ax.set_facecolor('black')        # Inner background

# Create a color map based on disturbance severity
norm = mcolors.Normalize(vmin=q1_df['Avg_Disturbance'].min(), vmax=q1_df['Avg_Disturbance'].max())
cmap = plt.cm.Reds
bar_colors = cmap(norm(q1_df['Avg_Disturbance']))

# Plot the bars (changed edgecolor to 'white' so dark red bars stand out against black)
ax.barh(q1_df['Source_Sheet'], q1_df['At_Risk_Count'], color=bar_colors, edgecolor='white')

# 4. Add and Style the Colorbar for Dark Mode
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax)
cbar.set_label('Average Disturbance Score (Darker = More Disturbance)', fontsize=12, color='white')

# Style colorbar ticks and outline to white
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white')
cbar.outline.set_edgecolor('white')

# 5. Make text, titles, and grid professional and White
plt.title('Vulnerability Matrix: At-Risk Populations & Disturbance by Park', fontsize=16, fontweight='bold', color='white')
plt.xlabel('Total At-Risk Birds Observed', fontsize=14, color='white')
plt.ylabel('National Park Code', fontsize=14, color='white')

# Set Ticks to White
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Set Borders (Spines) to White
for spine in ax.spines.values():
    spine.set_edgecolor('white')

# Set Grid lines to white and semi-transparent
ax.grid(axis='x', linestyle='--', alpha=0.3, color='white')

plt.tight_layout()
plt.show()
````
